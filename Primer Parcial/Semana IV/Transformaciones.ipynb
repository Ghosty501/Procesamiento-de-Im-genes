# Negativo

```python
import  numpy as np
```

```python
img_neg = np.zeros(img.shape, dtype=np.uint8)
```

```python
img_neg.dtype
```

```python
img.shape[0]
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img_neg[i, j] = 255 - img[i, j]
```

```python
plt.imshow(img_neg, cmap="gray")
plt.show()
```
