```python
histograma_g = cv.calcHist([img], [0], None, [256], (0, 256))
```

```python
histograma_r= cv.calcHist([img], [0], None, [256], (0, 256))
```

```python
plt.plot(histograma_r, "Red")
plt.show()
```

```python
colores = ["b", "g", "r"]
```

```python
for n, i in enumerate(colores):
    tmp = cv.calcHist([img], [n], None, [256], (0,256))
    plt.plot(tmp, i)
plt.show()
```

```python
for cuenta, valor in enumerate(colores):
    print(f"Cuenta = {cuenta} Valor= {valor}")
```
