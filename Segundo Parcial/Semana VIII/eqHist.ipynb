```python
import cv2 as cv 
import numpy as np 
from matplotlib import pyplot as plt
```

```python
img = cv.imread('f:/Imagenes/excavator-824655_1280.jpg', 0)
```

```python
plt.imshow(img, "gray")
plt.show()
```

```python

histograma = cv.calcHist([img], [0], None, [256], (0, 256))
plt.plot(histograma)
plt.show()
```

```python
# Ecualizacin

img_eq = cv.equalizeHist(img)
```

```python
plt.imshow(img_eq, "gray")
plt.show()
```

```python
plt.imshow(np.hstack((img, img_eq)), "gray")
plt.show()
```

```python
histograma_eq = cv.calcHist([img_eq], [0], None, [256], (0, 256))
plt.plot(histograma_eq)
plt.show()
```
