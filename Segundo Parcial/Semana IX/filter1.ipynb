```python
import cv2 as cv
import numpy as np 
from matplotlib import pyplot as plt
```

```python
img = cv.imread('f:/Imagenes/rubberwhale1.png', 0)
```

```python
img.shape
```

```python
img.size
```

```python
w = np.ones((3,3), dtype  = np.uint8)
```

```python
w = w * 1/9
```

```python
w.sum()
```

```python
img_filtered = cv.filter2D(img, -1, w)
```

```python
plt.imshow(np.hstack((img, img_filtered)), "gray")
plt.show()
```

```python
def gen_kernel(n):
    return np.ones((n,n), dtype = np.uint8)*1/n**2
```

```python
gen_kernel(9).sum()
```

```python
img_filtered = cv.filter2D(img, -1, gen_kernel(20))
plt.imshow(np.hstack((img, img_filtered)), "gray")
plt.show()
```
