```python
import cv2 as cv
import numpy as np 
from matplotlib import pyplot as plt
```

```python
sudoku = cv.imread('f:/Imagenes/sudoku.png', 0)
```

```python
plt.imshow(sudoku, "gray")
plt.axis("off")
plt.show()
```

----
# Umbralización global

```python
sudoku_bin_tradicional = cv.threshold(sudoku, 50, 255, cv.THRESH_BINARY)[1]
```

```python
plt.imshow(sudoku_bin_tradicional, "gray")
plt.axis("off")
plt.show()
```

---
# Umbralización adaptativa

```python
# Promedio
blocksize = 11
c= 2
sudoku_bin_adaptive = cv.adaptiveThreshold(sudoku, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, blocksize, c)
```

```python
plt.imshow(sudoku_bin_adaptive, "gray")
plt.axis("off")
plt.title(f"Blocksie = {blocksize} y C = {c}")
plt.show()
```

```python
# Gausiana

# Promedio
blocksize = 11
c= 2
sudoku_bin_adaptive = cv.adaptiveThreshold(sudoku, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, blocksize, c)
```

```python
plt.imshow(sudoku_bin_adaptive, "gray")
plt.axis("off")
plt.title(f"Blocksie = {blocksize} y C = {c}")
plt.show()
```

```python
# Umbral global, recomendado
plt.imshow(cv.threshold(sudoku, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)[1], "gray")
plt.show()
```

```python
babuino = cv.imread('f:/Imagenes/baboon.jpg', 0)
```

```python
th, img = cv.threshold(babuino, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
plt.imshow(img, "gray")
plt.title(f"Imagen con umbral{th}")
plt.show()
```

```python
# Umbral global, recomendado Triangle
th, img = cv.threshold(babuino, 0, 255, cv.THRESH_BINARY | cv.THRESH_TRIANGLE)
plt.imshow(img, "gray")
plt.title(f"Imagen con umbral{th}")
plt.show()
```
